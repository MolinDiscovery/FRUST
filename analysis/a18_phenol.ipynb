```python
conversion_factor =  627.509474

dimer_dft = -834847.294
H2_dft    = -737.1088023791428
HBpin_dft = -258305.91711374212

df = pd.read_parquet("../playground/results/phenol/phenol-job1111-251107-114139-e0/phenol.parquet")
G_phenol = (df["DFT-Opt-wB97X-D3-6-31G**-Freq-gibbs_energy"] - df["DFT-Opt-wB97X-D3-6-31G**-Freq-electronic_energy"]) + df["orca-wB97X-D3-6-31+G**-NoSym-electronic_energy"] # energy correction
G_phenol = G_phenol.iloc[0] * conversion_factor
G_phenol
```

# int2

```python
df = pd.read_parquet("../playground/results/phenol/int2-phenol-job1111-251109-215218-74/int2.parquet")
G = (df["DFT-Opt-wB97X-D3-6-31G**-Freq-gibbs_energy"] - df["DFT-Opt-wB97X-D3-6-31G**-Freq-electronic_energy"]) + df["orca-wB97X-D3-6-31+G**-NoSym-electronic_energy"] # energy correction
G = G.iloc[0] * conversion_factor
dG = (2 * (G - G_phenol) - dimer_dft) / 2
dG
```

```python
vis.plot_mols(df)
```

# mol2

```python
df = pd.read_parquet("../playground/results/phenol/mol2-phenol-job1111-251109-222310-d8/mol2.parquet")
G = (df["DFT-Opt-wB97X-D3-6-31G**-Freq-gibbs_energy"] - df["DFT-Opt-wB97X-D3-6-31G**-Freq-electronic_energy"]) + df["orca-wB97X-D3-6-31+G**-NoSym-electronic_energy"] # energy correction
G = G.iloc[0] * conversion_factor
dG = (2 * (G - G_phenol + H2_dft) - dimer_dft) / 2
dG
```

```python
vis.plot_mols(df)
```
