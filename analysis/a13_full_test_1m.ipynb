```python
import pandas as pd
import frust.vis as vis
```

```python
conversion_factor       =  627.509474
```

```python
dfts1 = pd.read_parquet("output_files/full_test_furan/ts1.parquet")
dfts2 = pd.read_parquet("output_files/full_test_furan/ts2.parquet")
dfts3 = pd.read_parquet("output_files/full_test_furan/ts3.parquet")
dfts4 = pd.read_parquet("output_files/full_test_furan/ts4.parquet")
dftint3 = pd.read_parquet("output_files/full_test_furan/int3.parquet")
dfmols = pd.read_parquet("output_files/full_test_furan/mols_full_test_furan.parquet")

dftint3.rename(columns={
    "DFT-wB97X-D3-6-31G**-Freq-electronic_energy": "DFT-wB97X-D3-6-31G**-OptTS-electronic_energy",
    "DFT-wB97X-D3-6-31G**-Freq-gibbs_energy": "DFT-wB97X-D3-6-31G**-OptTS-gibbs_energy",
    "DFT-wB97X-D3-6-31G**-Freq-normal_termination": "DFT-wB97X-D3-6-31G**-OptTS-normal_termination", 
}, inplace=True)

for df in [dfts1, dfts2, dfts3, dfts4, dftint3]:
    df.rename(columns={
        "DFT-pre-SP-wB97X-D3-6-31G**-NoSym-electronic_energy": "DFT-pre-SP",
        "DFT-pre-Opt-wB97X-D3-6-31G**-NoSym-electronic_energy": "DFT-pre-Opt",
        "DFT-wB97X-D3-6-31G**-OptTS-electronic_energy" : "DFT-Opt-EE",
        "DFT-wB97X-D3-6-31G**-OptTS-gibbs_energy": "DFT-Opt-G",
        "DFT-SP-wB97X-D3-6-31+G**-NoSym-electronic_energy": "DFT-SP-solv-EE"
    }, inplace=True)
    vib_cols = [col for col in df.columns if "vibs" in col or "coords" in col or "normal_termination" in col]
    df.drop(columns=vib_cols, inplace=True)

    df[[col for col in df if col.startswith("DFT")]] = df[[col for col in df if col.startswith("DFT")]] * conversion_factor

vis.plot_mols(dfmols, ligand_filter=["dimer"])
```

```python
cols_to_drop = [col for col in dfmols.columns if "normal_termination" in col or "vibs" in col or "coords" in col]
dfmols.drop(columns=cols_to_drop, inplace=True)
dfmols.rename(columns={
    "DFT-pre-SP-wB97X-D3-6-31G**-NoSym-electronic_energy": "DFT-pre-SP",
    "DFT-Opt-wB97X-D3-6-31G**-Freq-electronic_energy": "DFT-Opt-EE",
    "DFT-Opt-wB97X-D3-6-31G**-Freq-gibbs_energy": "DFT-Opt-G",
    "orca-wB97X-D3-6-31+G**-NoSym-electronic_energy": "DFT-SP-solv-EE",
}, inplace=True)

dfmols[[col for col in dfmols if col.startswith("DFT")]] = dfmols[[col for col in dfmols if col.startswith("DFT")]] * conversion_factor
dfmols
```

```python
dfmols["G_corr"] = (dfmols["DFT-Opt-G"] - dfmols["DFT-Opt-EE"]) + dfmols["DFT-SP-solv-EE"]
```

```python
dfts1["G_corr"] = (dfts1["DFT-Opt-G"] - dfts1["DFT-Opt-EE"]) + dfts1["DFT-SP-solv-EE"]
dfts2["G_corr"] = (dfts2["DFT-Opt-G"] - dfts2["DFT-Opt-EE"]) + dfts2["DFT-SP-solv-EE"]
dfts3["G_corr"] = (dfts3["DFT-Opt-G"] - dfts3["DFT-Opt-EE"]) + dfts3["DFT-SP-solv-EE"]
dfts4["G_corr"] = (dfts4["DFT-Opt-G"] - dfts4["DFT-Opt-EE"]) + dfts4["DFT-SP-solv-EE"]
dftint3["G_corr"] = (dftint3["DFT-Opt-G"] - dftint3["DFT-Opt-EE"]) + dftint3["DFT-SP-solv-EE"]
```

```python
dimer_G = dfmols[dfmols["ligand_name"] == "dimer"]["G_corr"].iloc[0]
print(f"{dimer_G = }")
catalyst_G = dfmols[dfmols["ligand_name"] == "catalyst"]["G_corr"].iloc[0]
HH_G = dfmols[dfmols["ligand_name"] == "HH"]["G_corr"].iloc[0]
ligand_G = dfmols[dfmols["ligand_name"] == "furan"]["G_corr"].iloc[0]

HBpin_lig_G_0 = dfmols[(dfmols["ligand_name"] == "HBpin-ligand") & (dfmols["rpos"] == 22.0)]["G_corr"].iloc[0]
HBpin_lig_G_1 = dfmols[(dfmols["ligand_name"] == "HBpin-ligand") & (dfmols["rpos"] == 23.0)]["G_corr"].iloc[0]

HBpin_mol_G = dfmols[dfmols["ligand_name"] == "HBpin-mol"]["G_corr"].iloc[0]

int2_0 = dfmols[(dfmols["ligand_name"] == "int2") & (dfmols["rpos"] == 17.0)]["G_corr"].iloc[0]
int2_1 = dfmols[(dfmols["ligand_name"] == "int2") & (dfmols["rpos"] == 18.0)]["G_corr"].iloc[0]
mol2_0 = dfmols[(dfmols["ligand_name"] == "mol2") & (dfmols["rpos"] == 17.0)]["G_corr"].iloc[0]
mol2_1 = dfmols[(dfmols["ligand_name"] == "mol2") & (dfmols["rpos"] == 18.0)]["G_corr"].iloc[0]


dE_cat = (2*catalyst_G - dimer_G) / 2
dE_TS1 = (2*(dfts1["G_corr"] - ligand_G) - dimer_G) / 2
dE_int2_0 = (2*(int2_0 - ligand_G + HH_G) - dimer_G) / 2
dE_int2_1 = (2*(int2_1 - ligand_G + HH_G) - dimer_G) / 2
dE_mol2_0 = (2*(mol2_0 - ligand_G + HH_G) - dimer_G) / 2
dE_mol2_1 = (2*(mol2_1 - ligand_G + HH_G) - dimer_G) / 2
dE_TS2 = (2*(dfts2["G_corr"] - ligand_G) - dimer_G) / 2
dE_TS3 = (2*(dfts3["G_corr"] - ligand_G - HBpin_mol_G + HH_G) - dimer_G) / 2
dE_TS4 = (2*(dfts4["G_corr"] - ligand_G - HBpin_mol_G + HH_G) - dimer_G) / 2
dE_int3 = (2*(dftint3["G_corr"] - ligand_G - HBpin_mol_G + HH_G) - dimer_G) / 2

import pandas as pd

vals = {
    "dE_cat": dE_cat,
    "dE_TS1": dE_TS1,
    "dE_int2_0": dE_int2_0,
    "dE_int2_1": dE_int2_1,
    "dE_mol2_0": dE_mol2_0,
    "dE_mol2_1": dE_mol2_1,
    "dE_TS2": dE_TS2,
    "dE_TS3": dE_TS3,
    "dE_TS4": dE_TS4,
    "dE_int3": dE_int3,
}

for k, v in vals.items():
    if isinstance(v, (pd.Series, pd.DataFrame)):
        # Example: show as list instead of raw Series
        print(f"{k}: {v.tolist()}")
    else:
        print(f"{k}: {v}")
```
