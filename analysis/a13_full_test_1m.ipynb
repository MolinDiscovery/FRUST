```python
import pandas as pd
import frust.vis as vis
```

```python
conversion_factor       =  627.509474
```

```python
dfts1 = pd.read_parquet("output_files/full_test_furan/ts1.parquet")
dfts2 = pd.read_parquet("output_files/full_test_furan/ts2.parquet")
dfts3 = pd.read_parquet("output_files/full_test_furan/ts3.parquet")
dfts4 = pd.read_parquet("output_files/full_test_furan/ts4.parquet")
dftint3 = pd.read_parquet("output_files/full_test_furan/int3.parquet")
dfmols = pd.read_parquet("output_files/full_test_furan/mols_full_test_furan.parquet")

dftint3.rename(columns={
    "DFT-wB97X-D3-6-31G**-Freq-electronic_energy": "DFT-wB97X-D3-6-31G**-OptTS-electronic_energy",
    "DFT-wB97X-D3-6-31G**-Freq-gibbs_energy": "DFT-wB97X-D3-6-31G**-OptTS-gibbs_energy",
    "DFT-wB97X-D3-6-31G**-Freq-normal_termination": "DFT-wB97X-D3-6-31G**-OptTS-normal_termination", 
}, inplace=True)

for df in [dfts1, dfts2, dfts3, dfts4, dftint3]:
    df.rename(columns={
        "DFT-pre-SP-wB97X-D3-6-31G**-NoSym-electronic_energy": "DFT-pre-SP",
        "DFT-pre-Opt-wB97X-D3-6-31G**-NoSym-electronic_energy": "DFT-pre-Opt",
        "DFT-wB97X-D3-6-31G**-OptTS-electronic_energy" : "DFT-Opt-EE",
        "DFT-wB97X-D3-6-31G**-OptTS-gibbs_energy": "DFT-Opt-G",
        "DFT-SP-wB97X-D3-6-31+G**-NoSym-electronic_energy": "DFT-SP-solv-EE"
    }, inplace=True)
    vib_cols = [col for col in df.columns if "vibs" in col or "coords" in col or "normal_termination" in col]
    df.drop(columns=vib_cols, inplace=True)

    df[[col for col in df if col.startswith("DFT")]] = df[[col for col in df if col.startswith("DFT")]] * conversion_factor

vis.plot_mols(dfmols, ligand_filter=["int2"])
vis.plot_mols(dfmols, ligand_filter=["mol2"])
```
