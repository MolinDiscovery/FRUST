# TS3 TMP with preoptimization first.

```python
df_ts3 = pd.read_parquet("../pipes/ts3_TMP_1/results_ts3_TMP.parquet")
summarize_ts_vibrations(df_ts3)
#vis.plot_vibs(df_ts3)
```

```python
print(df_ts3.columns)
atoms = df_ts3["atoms"].iloc[0]
coords = df_ts3["DFT-wB97X-D3-6-31G**-OptTS-opt_coords"].iloc[0]

xyz = cu.ac2xyz(atoms, coords, "ts3 TMP")
with open("../structures/ts3_TMP.xyz", "w") as f:
    f.write(xyz)
```

# TS4 TMP with preoptimization first.
