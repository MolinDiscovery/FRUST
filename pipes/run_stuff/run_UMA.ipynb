```python
from frust.pipes import run_mols_UMA
import pandas as pd
import os
```

```python
CSV_PATH = "../../datasets/1m.csv"
TS_XYZ         = "../structures/ts4_TMP.xyz"
df       = pd.read_csv(CSV_PATH)
smi_list = list(dict.fromkeys(df["smiles"]))
# job_inputs = create_ts_per_rpos(smi_list, TS_XYZ)
```

```python
out_dir = "UMA_test"

df = run_mols_UMA(
    smi_list,
    n_confs=None,
    n_cores=10,
    mem_gb=15,
    out_dir=out_dir,
    output_parquet=f"{out_dir}.parquet",
    DFT=False,
    save_output_dir=False,
    select_mols=["mol2"]
)
```

```python
import frust.vis as vis

vis.plot_mols(df, row_indices=[0], linked=True)
```
