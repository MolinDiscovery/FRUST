```python
from rdkit.Chem import rdDetermineBonds
from rdkit.Chem.rdchem import RWMol
from tooltoad.chemutils import xyz2mol
from tooltoad.vis import MolTo3DGrid
```

```python
with open("../../structures/ts2.xyz", "r") as f:
    inp_struct = f.read()

with open("../DFT-IRC-wB97X-D3-6-31G**-Freq/ts3_guess_0/input_IRC_F.xyz", "r") as f:
    irc_f_xyz = f.read()

with open("../DFT-IRC-wB97X-D3-6-31G**-Freq/ts3_guess_0/input_IRC_B.xyz", "r") as f:
    irc_b_xyz = f.read()

ts_i = RWMol(xyz2mol(inp_struct))
ts_f = RWMol(xyz2mol(irc_f_xyz))
ts_b = RWMol(xyz2mol(irc_b_xyz))
ts_i.RemoveBond(10, 12)
ts_i.RemoveBond(10, 41)

MolTo3DGrid(
    [ts_i, ts_b, ts_f],
    legends=["TS3", "TS3 IRC Backward", "TS3 IRC Forward"],
    cell_size=(450, 450),
    export_HTML="/Users/jacobmolinnielsen/Developer/FrustActivationProject/Presentations/01 Preliminary TS/assets/TS3_IRC.html",
)
```

```python
with open("../DFT-IRC-my-opt-wB97X-D3-6-31G**-Freq/ts3_font_IRC_F_0/input.xyz", "r") as f:
    opt_f_xyz = f.read()

with open("../DFT-IRC-my-opt-wB97X-D3-6-31G**-Freq/ts3_font_IRC_B_0/input.xyz", "r") as f:
    opt_b_xyz = f.read()

ts_opt_b = RWMol(xyz2mol(opt_b_xyz))
ts_opt_f = RWMol(xyz2mol(opt_f_xyz))
# export_HTML="/Users/jacobmolinnielsen/Developer/FrustActivationProject/Presentations/01 Preliminary TS/assets/TS1_IRC.html"
MolTo3DGrid(
    [ts_opt_b, ts_opt_f],
    legends=["TS1 IRC Backward opt", "TS1 IRC Forward Opt"],
    cell_size=(450, 450),
    export_HTML="/Users/jacobmolinnielsen/Developer/FrustActivationProject/Presentations/01 Preliminary TS/assets/TS3_IRC_opt.html",
)
```
