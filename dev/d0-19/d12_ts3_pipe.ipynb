```python
from frust.transformers import transformer_ts3
from frust.embedder import embed_ts
from frust.stepper import Stepper
from tooltoad.vis import MolTo3DGrid
from tooltoad.chemutils import ac2mol
```

```python
# smi_list = ["CC([Si](N1C=CC=C1)(C(C)C)C(C)C)C", "Cc1nc2cc(Br)ccc2o1"]
smi_list = ["CC([Si](N1C=CC=C1)(C(C)C)C(C)C)C"]
```

```python
ts_guess_struct="../structures/ts2.xyz"

ts_dict = {}
for i, smi in enumerate(smi_list):
    if i < 100:
        ts_mol = transformer_ts3(
            ligand_smiles=smi,
            ts_guess_struct=ts_guess_struct,
            embed_ready=True
        )
    
        ts_with_smi = {
            name: (mol, idxs, smi)
            for name, (mol, idxs) in ts_mol.items()
        }
        ts_dict.update(ts_with_smi)

embeds = embed_ts(ts_dict, ts_type="ts3", n_confs=5, optimize=True)

step = Stepper(smi_list, save_output_dir=False)
df0 = step.build_initial_df(embeds)
```

```python
df0
```

```python
df1 = step.xtb(df0, options={"gfnff": None, "opt": None}, constraint=True)
df2 = step.xtb(df1, options={"gfn": 2})

df2_filt = (
    df2.sort_values(["ligand_name", "rpos", "xtb-gfn-electronic_energy"])
        .groupby(["ligand_name", "rpos"])
        .head(2)
)
```

```python
df2_filt
```

```python
from rdkit.Chem.rdchem import RWMol
loc = 2

atoms = df2_filt["atoms"].iloc[loc]
coords_1 = df2_filt["coords_embedded"].iloc[loc]
coords_2 = df2_filt["xtb-gfnff-opt-opt_coords"].iloc[loc]

mol1 = ac2mol(atoms, coords_1)
mol2 = ac2mol(atoms, coords_2)

mol1_RW = RWMol(mol1)
mol2_RW = RWMol(mol2)

MolTo3DGrid([mol1_RW, mol2_RW], legends=['embed', 'xtb'], cell_size=(500,500), background_color="lightblue")
```
