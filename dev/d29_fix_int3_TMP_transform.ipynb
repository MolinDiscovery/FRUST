```python
from frust.transformers import transformer_int3
from frust.embedder import embed_ts
from frust.stepper import Stepper

ts_dict = transformer_int3(ts_guess_struct="../structures/int3_TMP.xyz")
embed_dict = embed_ts(ts_dict, n_confs=1)
```

```python
step = Stepper("C1=CC=CO1", step_type="int3")
df = step.build_initial_df(embed_dict)
#df = step.xtb(df, options={"gfnff": None, "opt": None}, constraint=True)
df = step.orca(df, options={"XTBFF": None, "Opt": None})
df
```

```python
import frust.vis as vis
import tooltoad.chemutils as cu

with open("../structures/int3.xyz") as f:
    xyzblock = f.read()
    mol1 = cu.xyz2mol(xyzblock)

with open("../structures/int3_TMP.xyz") as f:
    xyzblock = f.read()
    mol2 = cu.xyz2mol(xyzblock)

vis.MolTo3DGrid([mol1, mol2])
vis.plot_mols(df)
```
