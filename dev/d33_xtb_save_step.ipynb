```python
from frust.stepper import Stepper
from pathlib import Path
from tooltoad.chemutils import xyz2mol
from tooltoad.xtb import xtb_calculate
```

```python
# ## ORCA

f = Path("../structures/misc/TMPs/int3_TMP.xyz")
mols = {}
with open(f, "r") as file:
    xyz_block = file.read()
    mol = xyz2mol(xyz_block)
    mols[f.stem] = (mol, [0])

step = Stepper(list(mols.keys()), step_type="mol", save_output_dir=False)
df0 = step.build_initial_df(mols)

df = step.orca(df0, options={"XTB2": None, "sp": None}, save_step=False, save_files=None)
```

```python
## XTB

f = Path("../structures/misc/TMPs/int3_TMP.xyz")
mols = {}
with open(f, "r") as file:
    xyz_block = file.read()
    mol = xyz2mol(xyz_block)
    mols[f.stem] = (mol, [0])

step = Stepper(list(mols.keys()), step_type="mol", save_output_dir=False)
df0 = step.build_initial_df(mols)

df = step.xtb(df0, options={"gfn": 0, "opt": None}, save_step=False)
```

```python
df
```

```python
atoms = df["atoms"].iloc[0]
coords = df["coords_embedded"].iloc[0]

res = xtb_calculate(atoms, coords, options={"gfn": 1, "opt": None})
res
```
