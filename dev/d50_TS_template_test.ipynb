```python
import frust as ft
import pandas as pd
```

```python
TS = "TS2"
components = ft.screen.read("../datasets/font_smiles_cat.csv")
systems = ft.screen.expand(components)
ts_guesses = ft.screen.create_ts_guesses(
    systems,
    ts_types=[TS],
    n_confs=20,
    n_cores=10,
)
```

```python
step = ft.Stepper(n_cores=1)

df = step.xtb(
    ts_guesses[TS],
    name="xtb_preopt",
    options={"gfnff": None, "opt": None},
    constraint=True,
)
df = step.xtb(df, "xtb-gfn2-sp", {"gfn": 2})
df = ft.lowest_energy_rows(df)
df
```

```python
ft.plot_mols(df,range(0,len(df)), show_labels=False)
```
