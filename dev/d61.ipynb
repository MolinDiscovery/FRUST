```python
import frust as ft

components = ft.screen.read("../datasets/font_smiles_cat1.csv")
systems = ft.screen.expand(components)
ts_guesses = ft.screen.create_ts_guesses(
    systems,
    ts_types=["TS1", "TS2", "TS3", "TS4"],
    n_confs=3,
)
```

```python
systems
```

```python
ft.plot_mols(ts_guesses["TS3"])
```

```python
step = ft.Stepper()
df_ts3 = step.xtb(ts_guesses["TS3"], options={"gfnff": None, "opt": None}, constraint=True)
```
