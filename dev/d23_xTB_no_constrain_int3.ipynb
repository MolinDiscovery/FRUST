```python
from frust.pipes import run_ts_per_lig
import pandas as pd
```

```python
df = pd.read_csv("../datasets/1m.csv", index_col=0)
smiles = list(df["smiles"])
```

```python
df1 = run_ts_per_lig(
    smiles,
    ts_guess_xyz="../structures/int3_TMP.xyz",
    n_confs=1,
    n_cores=5,
    save_output_dir=False,
)
```

```python
df1
```

```python
from tooltoad.chemutils import ac2mol
from tooltoad.vis import MolTo3DGrid

idx = 28 # working 12

atoms = df1["atoms"].iloc[idx]

coords1 = df1["coords_embedded"].iloc[idx]
coords2 = df1["xtb-gfnff-opt-opt_coords"].iloc[idx]
coords3 = df1["xtb-gfn-opt-opt_coords"].iloc[idx]

all_coords = [coords1, coords2, coords3]
all_mols = [ac2mol(atoms, c) for c in all_coords]

print(df1["ligand_name"].iloc[idx], "rpos:", df1["rpos"].iloc[idx])
MolTo3DGrid(all_mols, legends=['embed', 'xtb-ff', 'xtb-opt'])
```
