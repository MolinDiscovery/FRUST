### DFT Freqs on UMA optimized structure (UMA small)

````python
import frust.vis as vis
from frust.utils.analytics import summarize_ts_vibrations
from tooltoad.vis import show_vibs

df1 = pd.read_parquet("../ts4-uma-freq/UMA-opt-UMA-freqs.parquet")

i = 0
results_dict = {
    "atoms": list(df1["atoms"].iloc[i]),
    "opt_coords": list(df1["UMA-test-1-OptTS-ExtOpt-opt_coords"].iloc[i]),
    "vibs": list(df1["UMA-freq-ExtOpt-NumFreq-vibs"].iloc[i]),
}

summarize_ts_vibrations(df1, "UMA-freq-ExtOpt-NumFreq-vibs")
print("")
show_vibs(results_dict, numFrames=40, vId=0, fps=10)
````
