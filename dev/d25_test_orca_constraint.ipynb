```python
from frust.stepper import Stepper
import frust.vis as vis
import pandas as pd
```

```python
step = Stepper(["CO"], step_type="TS3", save_output_dir=False)

df = pd.read_parquet("../analysis/output_files/04_ts3_TMP.parquet")
df = step.orca(df.iloc[[0]], options={"XTB2": None, "Opt": None}, save_files=None, constraint=True)
```

```python
vis.plot_mols(df)
```
