```python
import frust as ft
import pandas as pd
```

```python
method = ft.workflows.methods.preset("r2scan-3c")

wf = ft.workflows.screen_ts(
    csv_path="../datasets/font_smiles_cat1.csv",
    ts_types=["TS1", "TS2", "TS3", "TS4"],
    method=method,
    n_confs=None,
    top_n=10,
    dft=False,
)

ts_guesses = ft.screen.create_ts_guesses(
    ft.screen.expand(ft.screen.read("../datasets/font_smiles_cat1.csv")),
    ts_types=["TS3"],
    n_confs=5,
)
```

```python
ts = ts_guesses["TS3"].reset_index(drop=True)

ft.plot_mols(
    ts[ts["rpos"] == 0],
    include_coords=["coords_embedded"],
    columns=5,
    cell_size=(400, 400),
    show_charges=False,
)
```

```python
csv_path="../datasets/font_smiles_cat1.csv"

components = ft.screen.read(csv_path)
components
```

```python
systems = ft.screen.expand(components)
systems
```

```python
ts_type = "TS3"
ts_guesses = ft.screen.create_ts_guesses(
    systems,
    ts_types=[ts_type],
    n_confs=5,
)

df_ts = ts_guesses[ts_type][ts_guesses.get(ts_type)["rpos"] == 0]
```

```python
ft.plot_mols(df_ts, columns=5, cell_size=(400,400))
```

```python
ts_type = "TS3"
ts_guesses = ft.screen.create_ts_guesses(
    systems,
    ts_types=[ts_type],
    n_confs=20,
)

df_ts = ts_guesses[ts_type][ts_guesses.get(ts_type)["rpos"] == 0]
df_ts
step = ft.Stepper()

df_ts = step.xtb(
    df_ts,
    name="xtb_preopt",
    options={"gfn": 0, "opt": None},
    constraint=True,
    n_cores=1,
)
```

```python
ft.plot_mols(df_ts)
```

```python
ft.plot_mols(ft.lowest_energy_rows(df_ts, n=5))
```
